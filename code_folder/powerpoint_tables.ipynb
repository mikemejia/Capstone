```python
races = set(list(data[(data.circuitId) & (data.target == True) & (data.date == 2019)]['circuitId']))
```

```python
races
```

```python
places =['Australia', 'Bahrain', 'Canada', 'UK', 'Hungary', 'Belgium',
        'Singapore', 'China', 'USA', 'Russia', 'Azerbaijan']
```

```python
year = [2015, 2016,2017,2018,2019]
year_s = ['2015','2016','2017','2018','2019']
```

```python
avg_races = []
for i in races:
    avg_races.append(data[(data.circuitId == i) & (data.target == True) & (data.date == 2019)].shape[0])
```

```python
avg_races
```

```python
_1=pd.Series(avg_races).pct_change()
```

```python
_2=pd.Series(avg_races).pct_change()
```

```python
_3=pd.Series(avg_races).pct_change()
```

```python
plt.figure(figsize = (10,10))
plt.plot(year_s,_1, label = "Australia");
plt.plot(year_s,_2, label = "Bahrain");
plt.plot(year_s,_3, label = "Canada");
plt.title('Change in pitstops', size = 25);
plt.ylabel('Change %', size = 20);
plt.xlabel('Year', size = 20)
plt.legend(loc="upper right")
plt.legend(fontsize=20)
plt.xticks(size = 30)
plt.yticks(size = 20);
```
